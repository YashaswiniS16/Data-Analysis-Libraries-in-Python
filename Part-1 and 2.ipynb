## Classes and Objects

```python
class Employee:
    id = 0
    name = ''
    gender = ''
    age = ''
    
    def __init__(self,empid,name,gender,age):
        self.id=empid
        self.name=name
        self.gender=gender
        self.age=age
    
    def ShowEmpInfo(self):
        print("Employee Id:",self.id)
        print("Employee Name:",self.name)
        print("Employee Gender:",self.gender)
        print("Employee Age:",self.age)
```

# Window Functions

periodic data: it is nothing but a data which is based upon the dates
    
so window functions is used to analze the periodic data

window function has 2 types:
   
   1. Rolling
   2. Expanding

```python
import pandas as pd
import numpy as np
```

```python
price=np.random.randn(10,1)
days=pd.date_range('1/1/2010',periods=10)
```

```python
days
```

```python
price
```

```python
dfdp=pd.DataFrame(price,index=days,columns=['StockPrice'])
dfdp
```

```python
#To check prevoius day's price
dfdp["prev_price"]=dfdp["StockPrice"].shift(1)                               
```

```python
dfdp
```

```python
# Now to calculate the changes
dfdp["Daily_returns"]=dfdp["StockPrice"]/dfdp['prev_price']-1
dfdp
```

###### Expanding mean

```python
dfdp['expand_mean']=dfdp['Daily_returns'].expanding().mean()
dfdp
```

###### Rolling


```python
dfdp['rool_mean_3']=dfdp["Daily_returns"].rolling(3).mean()
dfdp
```

```python
df1.join(df2,lsuffix='df1',rsuffix='df2',how='left')
```

###### Right join

```python
df1.join(df2,lsuffix='df1',rsuffix='df2',how='right')
```

###### Inner join

```python
df1.join(df2,lsuffix='df1',rsuffix='df2',how='inner')
```

###### outer
